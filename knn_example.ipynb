```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score, classification_report
```

```python
#read the breast cancer dataset
df = pd.read_csv('breast-cancer.csv')
df.head()
```

```python
#replace diagnosis M for Malignant with 1 i.e cancerous and Benign B for 0 i.e not cancerous
df['diagnosis'] = df['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)

df.head()
```

```python
#split the df to X i.e features and y i.e output
X = df[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean', 'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']]
y = df['diagnosis']

#to avoid any bias reshape
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
```

```python
#split our dataset for the training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1)
```

```python
#init our model using K Nearest Neighbors Model for Classification
knn_model = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
```

```python
#train our model using K Nearest Neighbors Model for Classification
knn_model.fit(X_train, y_train)
```

```python
#predict our model using K Nearest Neighbors Model for Classification
y_pred = knn_model.predict(X_test)
```

```python
#check for performance model using K Nearest Neighbors Model for Classification
performance = classification_report(y_test, y_pred)
print(performance)
print(confusion_matrix(y_test, y_pred))
#knn model works super great.
```

```python
def generateNewDf(cols, y_list, X_list):
  column_names = cols.tolist()
  y_col = column_names[0]
  X_cols = column_names[1:]
  print('Y COLUMN', y_col)
  print('X COLUMN', X_cols)

  # Create a dictionary with column names and values
  data_dict = {y_col: y_list.tolist()}

  # Convert the X_test NumPy array to a DataFrame with appropriate column names
  X_test_df = pd.DataFrame(X_list, columns=X_cols)

  # Iterate over X_cols and add each column to the data_dict
  for col in X_cols:
      data_dict[col] = X_test_df[col].tolist()

  # Create the result_df DataFrame
  result_df = pd.DataFrame(data_dict)
  return result_df
```

```python
#generate a result based on prediction and the actual dataset
pred_dataset = generateNewDf(df.columns, y_pred, X_test )
pred_dataset.to_csv('predicted-breast canver dataset.csv', index=False)
actual_dataset = generateNewDf(df.columns, y, X )
actual_dataset.to_csv('actual-breast canver dataset.csv', index=False)
```
